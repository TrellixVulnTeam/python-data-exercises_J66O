In this exercise we will be exploring and analyzing a dataset of Apps in the Google Play store.

```python
import numpy as np
import pandas as pd

original_df = pd.read_csv("./googleplaystore.csv")
df = original_df.copy()
```

### Plot the number of apps in each Category

```python
categories = df.groupby('Category').agg({'App' : 'size'})
categories.sort_values(by='App', ascending=False).plot(kind='bar', figsize=(20,10))
```

# what are the top 10 rated apps in each category?

hint: sort and by category and rating begore you group. 

```python
categories = df.loc[:, ['Category', 'App', 'Rating']]
categories.sort_values(['Category','Rating'], ascending=False).groupby('Category').head(10)
```

# How many Installs are there in each category? 


```python
df = original_df.copy()
def clean_number(x):
    x = x.replace(',', '')
    x = x.replace('+', '')    
    return  float(x) if x.isdigit()  else 0

df['Installs'] = df['Installs'].apply(clean_number)
df.groupby('Category').agg({'Installs': 'sum'}).sort_values(by='Installs').plot(kind='bar', figsize=(20,10))

```

# Does the size of the app correlates with its Rating? In every category?

hint: make sure to clean and format your data

```python
import matplotlib.pyplot as plt

rat_vs_size = df.loc[:, ['Category', 'Size', 'Rating']]
rat_vs_size['Size'] = rat_vs_size['Size'].apply(lambda x: float(x.replace('M', '')) if x.replace('M', '').isnumeric() else np.nan )
res = rat_vs_size.groupby('Category').corr()

res = res.drop(index='Rating', level=1)
res.drop(columns='Size').sort_values(by='Rating', ascending=False)

```
