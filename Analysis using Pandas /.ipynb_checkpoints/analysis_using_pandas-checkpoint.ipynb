# Data Analysis using Pandas


```python
import numpy as np
import pandas as pd

original_df = pd.read_csv("./soccerdb/Match.csv")
df = original_df.copy()
```

# Which agenncy gave the best odds for the results that actually happend?

Each betting agency provides betting odds for a home team win, draw, and away team win. 

Only one of those outcomes actually happened. 

Lets check which agency gave the highest reward for the actuall winning results over time, on average. 




### Calculate the actuall games results

```python
# add columns to the dataset and set the result type
home_win = (df['home_team_goal'] > df['away_team_goal'])
away_win = (df['home_team_goal'] < df['away_team_goal'])
draw = (df['home_team_goal'] == df['away_team_goal'])

df.loc[home_win, 'home_win'] = 1  
df.loc[~home_win, 'home_win'] = 0
df.loc[away_win, 'away_win'] = 1  
df.loc[~away_win, 'away_win'] = 0
df.loc[draw, 'draw'] = 1  
df.loc[~draw, 'draw'] = 0
```

### Set the values to numric


```python
betting_columns = df.columns[-30:-3]    
    
for column in betting_columns:
    df.loc[:,column] = pd.to_numeric(df[column], errors='coerce')

betting_ods.replace("<null>", 0, inplace=True)
```

### Iter over betting agencies and zero out the false reuslts 

```python

for h, d, a in zip(*[iter(betting_columns)]*3):
    df[h] *= df['home_win']
    df[d] *= df['draw']
    df[a] *= df['away_win']

    
```

### Count the number of times each agency gave the best odd for a game


```python
# use the idmax function to get the index of the max value in each row 
# https://pandas.pydata.org/pandas-docs/version/0.17.0/generated/pandas.DataFrame.idxmax.html

betting_ods = df[betting_columns]                
win_count = betting_ods.idxmax(axis=1).value_counts()
win_count.index = win_count.index.map(lambda x: x[:2])
res = win_count.groupby(win_count.index).agg(np.sum).sort_values(ascending=False)
```

### Plot a bar chart of the result

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.bar(res.index, res)
plt.xlabel('Betting Agency', fontsize=12)
plt.ylabel('Best odd', fontsize=12)
plt.xticks(res.index, res.index, fontsize=15)
plt.title('Betting Agency by # of best returns for winning', fontsize=20)
plt.show()

# Which agency gives the best offer.
# which agency is best at predicting the result max(h/d/a) - min(h/d/a)

# https://www.bestbettingsites.uk/#testing
```

```python
# Convert decimals odds into a fraction by subtracting 1, and using 1 as the denominator.
# Example: 3.40 â€“ 1 = 2.40.  This creates the decimal odds of 2.40/1.

df = original_df.copy()
# reset the data frame


betting_columns = df.columns[-30:-3]    
betting_ods = df[betting_columns]

for column in betting_columns:
    betting_ods.loc[:,column] = pd.to_numeric(betting_ods[column], errors='coerce')
    betting_ods[column].fillna(0)

betting_ods = betting_ods.apply(lambda x: 1/(x + 1 - 1)) # we need to remove 1 to convert to fractional odds, and add 1 to calcualte the odds level

agencies = []
for h, d, a in zip(*[iter(betting_columns)]*3):
    agency = h[:2]
    agencies.append(agency)
    betting_ods[agency] = 1 / (betting_ods[h] + betting_ods[d] + betting_ods[a])
    
margins = betting_ods[agencies].apply(lambda x: 1-x)
margins['season'] = df['season']
margins = margins.groupby('season').agg(['mean']) #.transpose()
margins.applymap(lambda x: x*100)

import matplotlib.ticker as mtick

fig, ax = plt.subplots()
ax.legend(loc='best')
ax.yaxis.set_major_formatter(mtick.PercentFormatter(decimals=2))
margins.plot(figsize=(20,10), ax=ax)





```
