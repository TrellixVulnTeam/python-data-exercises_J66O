## Who are the top 10 teams with the most goals ?


```python
match_score = match[['home_team_api_id',  
                     'home_team_goal', 
                     'away_team_api_id', 
                     'away_team_goal']]
match_score.head()
```

```python
# To get the teams' names, we'll need to join the Team table
teams = pd.read_csv("./soccerdb/team.csv")
teams.head()
```

```python
home_teams = teams[['team_api_id', 'team_long_name']]
home_teams = home_teams.add_prefix('home_')
match_score = pd.merge(left=match_score, 
                       right=home_teams, 
                       on='home_team_api_id', 
                       how='left')
match_score.head()
```

```python
# now merge another projection of Team with an away prefix
away_teams = teams[['team_api_id', 'team_long_name']]
away_teams = away_teams.add_prefix('away_')
match_score = pd.merge(left=match_score, 
                       right=away_teams, 
                       on='away_team_api_id', 
                       how='left')
match_score.head()
```

```python
# lets reorder the columns and see what we've got
home_score = match_score[['home_team_long_name', 'home_team_goal']] 
away_score = match_score[['away_team_long_name', 'away_team_goal']]
print(away_score.head())

print('\n')
print(home_score.head())
```

```python
away_score_total = away_score.groupby('away_team_long_name').agg(np.sum)
home_score_total = home_score.groupby('home_team_long_name').agg(np.sum)

away_score_total.rename(columns={'away_team_goal' : 'goals'}, inplace=True)
home_score_total.rename(columns={'home_team_goal' : 'goals'}, inplace=True)

all_totals = pd.concat([away_score_total, home_score_total])
teams_total_score = all_totals.groupby(all_totals.index).agg(np.sum).sort_values(by='goals', ascending=False)[:10]
teams_total_score
```

```python
import matplotlib.pyplot as plt

height = teams_total_score.index
bars = teams_total_score.goals
y_pos = np.arange(len(bars))

plt.barh(height,bars)
plt.xticks(rotation='vertical')

for a, b in zip(bars, height):
    plt.text(a, b, str(a), color='g')

plt.show()
```
